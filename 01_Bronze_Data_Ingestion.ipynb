```python
# Read raw e-commerce data
raw_df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/default/datastore/online_retail.csv")
)

raw_df.display()
```

```python
from pyspark.sql.functions import current_timestamp

bronze_df = (
    raw_df
    .withColumn("ingestion_timestamp", current_timestamp())
)
```

```python
%sql
CREATE DATABASE IF NOT EXISTS ecommerce_ai;
```

```python
%sql
USE ecommerce_ai;
```

```python
bronze_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("ecommerce_ai.bronze_online_retail")
```

```python
%sql
SELECT 
    COUNT(*) AS total_records,
    COUNT(DISTINCT CustomerID) AS unique_customers
FROM ecommerce_ai.bronze_online_retail;
```

```python
%sql
DESCRIBE TABLE ecommerce_ai.bronze_online_retail;
```
